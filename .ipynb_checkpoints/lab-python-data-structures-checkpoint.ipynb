## Exercise 1: Working with Lists

Imagine you are building a program for a teacher who wants to track the progress of their students throughout the semester. The teacher wants to input the grades of each student one by one, and get a summary of their performance. There are in total 5 students. You are tasked with building the program that will allow the teacher to do this easily.

The program will prompt the teacher to enter the grades of each student. Once the teacher has entered all the grades, the program will calculate the total sum of the grades and display it on the screen. Then, the program will create a new list by selecting only the grades of the first, third, and fifth students entered by the teacher, and sort them in ascending order.

Finally, the program will print out the new list, along with its length and the number of occurrences of the score 5 in the list. This will give the teacher a good overview of the performance of the selected students, and help them identify any potential issues early on.

*Hint:*
- You can use the input() function to ask the user to enter their information.
- Look for list methods to perform the tasks. 
- Remember, it is possible to get a part of the sequence using:

    ```python
    sequence[x:y:z]
    ```
    where x, y, z are integers.

    The above returns a new sequence with the following characteristics:

    - A sequence with the same type as the original (a slice of a list is a list, a slice of a tuple is a tuple, and a slice of a string is a string).
    - A sequence with elements from `sequence [x]` to `sequence [y-1]` (does not include a sequence [y]). By skipping `z` elements each time, it can be omitted if ` z = 1`.


*Recommended External Resources:*
- *[Python Lists](https://www.w3schools.com/python/python_lists.asp)*
- *[Python List Methods](https://www.w3schools.com/python/python_ref_list.asp)*
- *[Python Built-in Functions](https://docs.python.org/3/library/functions.html)*


````python
grades=[]
for i in range(5):
    grade=int(input("Enter student grades (one by one):"))
    grades.append(grade)

total=sum(grades)
print(f"The total sum of grades is {total}")

new_list=grades[::2]
new_list.sort()

print(f"The new list of the first 3 grades is {new_list}")
print(f"The lengh of the new list is {len(new_list)}")
print(f"The number of occurrencies of a grade of 5 was {(new_list.count(5))}")
````

## Exercise 2: Tuples

Imagine you're running a fruit stand and want to keep track of your inventory. Write a Python program that does the following:

- Initializes a tuple with 5 different types of fruit.
- Outputs the first and last elements of the tuple, so you can see the full range of fruits the store offers.
- Replaces the second element of the tuple with a new fruit that the store has recently received, and prints the updated tuple so you can see the changes.
- Concatenates a new tuple containing 2 additional fruits to the original tuple, so you can add them to the store inventory, and prints the resulting tuple to see the updated inventory.
- Splits the resulting tuple into 2 tuples of 3 elements each (the first tuple contains the first 3 elements, and the second tuple contains the last 3 elements), so you can organize the inventory more effectively.
- Combines the 2 tuples from the previous step with the original tuple into a new tuple, and prints the resulting tuple and its length, so you can see the final inventory after all the changes.

*Recommended External Resources: [Python Tuples Examples and Methods](https://www.w3schools.com/python/python_tuples.asp)*



````python
fruits=("grape","banana","orange","apple","strawberry")
print("The first fruit is ",fruits[0])
print("The last fruit is ",fruits[-1])

#Replace (Tuples can't be modified -> turn into a list and back again to a Tuple)
fruit_list=list(fruits)
fruit_list[1]="peach"
fruits=tuple(fruit_list)
print(fruits)

#Concatenate
new_fruits=("mango","kiwi","pineapple")
fruits=fruits+new_fruits
print(fruits)

#Split Turples
tuple1=fruits[:3]
tuple2=fruits[-3:]

print("The first Tuple has these fruits:",tuple1)
print("The second Tuple has these fruits:",tuple2)

#New mega tuple
mega_tuple=tuple1+tuple2
print("The final tuple has these fruits:",mega_tuple)
print("The lengh of this last tuple in therms of fruits is:",len(mega_tuple))
````

## Exercise 3: Sets

Imagine you are a data analyst working for a literature museum. Your manager has given you two poems to analyze, and she wants you to write a Python program to extract useful information from them.

Your program should:

- Create two sets, one for each poem, containing all unique words in both poems (ignoring case and punctuation).
- Print the number of unique words in each set.
- Identify and print the unique words present in the first poem but not in the second one.
- Identify and print the unique words present in the second poem but not in the first one.
- Identify and print the unique words present in both poems and print it in alphabetical order.

*Recommended External Resources:*
- *[Python Sets](https://www.w3schools.com/python/python_sets.asp)* 
- *[Python Set Methods](https://www.w3schools.com/python/python_ref_set.asp)*
- *[Python String Methods](https://www.w3schools.com/python/python_ref_string.asp)*


````python
poem = """Some say the world will end in fire,
Some say in ice.
From what I’ve tasted of desire
I hold with those who favor fire.
But if it had to perish twice,
I think I know enough of hate
To say that for destruction ice
Is also great
And would suffice."""

new_poem = """Some say life is but a dream,
Some say it's a test.
From what I've seen and what I deem,
I side with those who see it as a quest.

But if it had to end today,
I think I know enough of love,
To say that though it fades away,
It's still what we are made of."""
````

````python
# Create two sets of words
clean_poem= poem.lower().replace(".","").replace(",","").replace("’","").replace("'","")
clean_new_poem= new_poem.lower().replace(".","").replace(",","").replace("’","").replace("'","")

#Make sure to have unique values
poem_list = clean_poem.split()
new_poem_list= clean_new_poem.split()

poem_set=set(poem_list)
new_poem_set=set(new_poem_list)

print("The unique words of the first poem are: ", poem_list)
print("The unique words of the second poem are: ", new_poem_list)

#Print list lengh
print("The number of unique words in the first poem is ",len(poem_set))
print("The number of unique words in the second poem is ",len(new_poem_set))

#Identify and print the unique words present in the first poem 
print("Only in first poem: ", poem_set - new_poem_set)

#Identify and print the unique words present in the second poem
print("Only in second poem: ", new_poem_set - poem_set)

#Identify and print the unique words present in both poems + soting
both_poems_words=poem_set.intersection(new_poem_set)
print("Common words in both poems: ", sorted(both_poems_words))
````

## Exercise 4: Dictionaries

Consider the following dictionary of students with their scores in different subjects. One of the students, Bob, has complained about his score in Philosophy and, after reviewing it, the teacher has decided to update his score to 100. Write a Python program that updates Bob's score in Philosophy to 100 in the dictionary.

*Recommended External Resources: [Python Dictionary Examples and Methods](https://www.w3schools.com/python/python_dictionaries.asp)*


````python
grades = {'Alice': {'Physics': 75, 'Math': 85, 'Chemistry': 60, 'Philosophy': 90}, 
          'Bob': {'Physics': 75, 'Math': 85, 'Chemistry': 60, 'Philosophy': 90}}
````

````python
grades['Bob']['Philosophy']=100
print(grades)
````

## Bonus

1. Below are the two lists. Write a Python program to convert them into a dictionary in a way that item from list1 is the key and item from list2 is the value.

*Hint: Use the zip() function. This function takes two or more iterables (like list, dict, string), aggregates them in a tuple, and returns it. Afterwards, you can use a function that turns a tuple into a dictionary.*

*Recommended External Resources: [Python Zip Function](https://www.w3schools.com/python/ref_func_zip.asp)*


````python
keys = ['Physics', 'Math', 'Chemistry', 'Philosophy']
values = [75, 85, 60,90]

# Dictionary conversion
zip(keys,values)
new_dict=dict(zip(keys,values))

print(new_dict)
````

2. Get the subject with the minimum score from the previous dictionary.

*Hint: Use the built-in function min(). Read about the parameter key.*


*Recommended External Resources:*
- *[Python Min Function Official Documentation](https://docs.python.org/3.8/library/functions.html#min)*
- *[How to use key function in max and min in Python](https://medium.com/analytics-vidhya/how-to-use-key-function-in-max-and-min-in-python-1fdbd661c59c)*

````python
# Minimum score subject

min_score=min(new_dict,key=new_dict.get)
print("The subject with the minimum score was:", min_score)
````
